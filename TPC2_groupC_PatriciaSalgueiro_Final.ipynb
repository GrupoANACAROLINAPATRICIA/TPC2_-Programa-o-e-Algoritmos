```python
#primeiro ler o csv
import pandas as pd
df_datasales = pd.read_csv("company_sales_data.csv")
df_datasales.head()
```

```python
# criar variavel profitlist e monthlist, tendo em conta que tolist fara com os dados estejam guardados numa lista
profitList = df_datasales ['total_profit'].tolist()
monthList = df_datasales ['month_number'].tolist()
```

```python
#importar a biblioteca matplotlib
import matplotlib.pyplot as plt
```

```python
# aplicar a função em x terá a monthlist e o y a profitlist
matplotlib.pyplot.plot(monthList,profitList)
matplotlib.pyplot.xlabel('Month Number')
matplotlib.pyplot.ylabel('Total profit')


matplotlib.pyplot.show()
```

```python
#Definição da label com caraterização da linha
profit_data= matplotlib.pyplot.plot(monthList,profitList,"r-", linestyle= 'dotted', lw=3.0, marker='o',markerfacecolor='black')

#caraterização do gráfico
matplotlib.pyplot.xlabel('Número do mês')
matplotlib.pyplot.ylabel('Número de unidades vendidas')

#Caraterização da legenda
plt.legend(profit_data,['Profit of last year'],loc=4)
plt.title ("Company Sales data of Last Year")


matplotlib.pyplot.show()
```
